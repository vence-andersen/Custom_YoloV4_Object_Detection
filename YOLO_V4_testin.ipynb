```python
from ctypes import *
import random
import os
import cv2
import time
import darknet
import argparse
from threading import Thread, enumerate
from queue import Queue
from google.colab.patches import cv2_imshow
import face_recognition
import PIL
import numpy as np
import copy


def set_saved_video(input_video, output_video, size):
    fourcc = cv2.VideoWriter_fourcc(*"MJPG")
    fps = int(input_video.get(cv2.CAP_PROP_FPS))
    video = cv2.VideoWriter(output_video, fourcc, fps, size)
    return video


def convert2relative(bbox):
    """
    YOLO format use relative coordinates for annotation
    """
    x, y, w, h  = bbox
    _height     = darknet_height
    _width      = darknet_width
    return x/_width, y/_height, w/_width, h/_height


def convert2original(image, bbox):
    x, y, w, h = convert2relative(bbox)

    image_h, image_w, __ = image.shape

    orig_x       = int(x * image_w)
    orig_y       = int(y * image_h)
    orig_width   = int(w * image_w)
    orig_height  = int(h * image_h)

    bbox_converted = (orig_x, orig_y, orig_width, orig_height)

    return bbox_converted


def convert4cropping(image, bbox):
    x, y, w, h = convert2relative(bbox)

    image_h, image_w, __ = image.shape

    orig_left    = int((x - w / 2.) * image_w)
    orig_right   = int((x + w / 2.) * image_w)
    orig_top     = int((y - h / 2.) * image_h)
    orig_bottom  = int((y + h / 2.) * image_h)

    if (orig_left < 0): orig_left = 0
    if (orig_right > image_w - 1): orig_right = image_w - 1
    if (orig_top < 0): orig_top = 0
    if (orig_bottom > image_h - 1): orig_bottom = image_h - 1

    bbox_cropping = (orig_left, orig_top, orig_right, orig_bottom)

    return bbox_cropping


def video_capture(frame_queue, darknet_image_queue):
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print('Oh! Hello')
            break
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame_resized = cv2.resize(frame_rgb, (darknet_width, darknet_height),
                                   interpolation=cv2.INTER_LINEAR)
        frame_queue.put(frame)
        img_for_detect = darknet.make_image(darknet_width, darknet_height, 3)
        darknet.copy_image_from_bytes(img_for_detect, frame_resized.tobytes())
        darknet_image_queue.put(img_for_detect)
    cap.release()


def inference(darknet_image_queue, detections_queue, fps_queue):
    while cap.isOpened():
        darknet_image = darknet_image_queue.get()
        prev_time = time.time()
        detections = darknet.detect_image(network, class_names, darknet_image, thresh=0.75)
        detections_queue.put(detections)
        fps = int(1/(time.time() - prev_time))
        fps_queue.put(fps)
        # print("FPS: {}".format(fps))
        # darknet.print_detections(detections, coordinates=False)
        darknet.free_image(darknet_image)
    cap.release()


def bbox2points(bbox):
    """
    From bounding box yolo format
    to corner points cv2 rectangle
    """
    x, y, w, h = bbox
    xmin = int(round(x - (w / 2)))
    xmax = int(round(x + (w / 2)))
    ymin = int(round(y - (h / 2)))
    ymax = int(round(y + (h / 2)))
    return xmin, ymin, xmax, ymax


def draw_boxes(detections, image, colors):
    faces = []
    frame = copy.deepcopy(image)
    for label, confidence, bbox in detections:
        left, top, right, bottom = bbox2points(bbox)
        faces.append(frame[top:bottom, left:right])
        cv2.rectangle(image, (left, top), (right, bottom), colors[label], 1)
        cv2.putText(image, "{} [{:.2f}]".format(label, float(confidence)),
                    (left, top - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                    colors[label], 2)
    return image, faces


known_faces = [];all_faces = []; Remember = set(); so_far = set()
def drawing(frame_queue, detections_queue, fps_queue):
    random.seed(3)  # deterministic bbox colors
    global known_faces, all_faces, Remember, so_far, w
    while cap.isOpened():
        frame = frame_queue.get()
        detections = detections_queue.get()
        fps = fps_queue.get()
        detections_adjusted = []
        if frame is not None:
            for label, confidence, bbox in detections:
                if str(label).lower() == 'face':           
                    bbox_adjusted = convert2original(frame, bbox)
                    detections_adjusted.append((str(label), confidence, bbox_adjusted))
                else:
                    Remember.add(label)
            image, faces = draw_boxes(detections_adjusted, frame, class_colors)
            cur_frame = []
            no_of_faces = len(faces)
            so_far.add(no_of_faces)
            for face in faces:
                if face.shape[0] < 20 or face.shape[1] < 20:
                    continue
                im = PIL.Image.fromarray(face)
                im = im.convert('RGB')
                im = np.array(im)
                height_1, width_1, _ = im.shape
                face_location_1 = (0, width_1, height_1, 0)
                encodings_1 = face_recognition.face_encodings(im, known_face_locations=[face_location_1], model='large')[0]
                cur_frame.append(encodings_1)
                if not known_faces:
                    known_faces.append(encodings_1)
                    all_faces.append(encodings_1)
                    cv2.imwrite(f'/content/spl/{time.time()}.jpg', face)
                results_1 = face_recognition.compare_faces(known_faces, encodings_1, tolerance=0.6)
                if True not in results_1:
                    try:
                        dist = face_recognition.face_distance(pre_frame, encodings_1)
                        if not any(x < 0.65 for x in dist):
                            known_faces.append(encodings_1)
                            cv2.imwrite(f'/content/spl/{time.time()}.jpg', face)
                    except:
                        continue
            if max(so_far) > len(known_faces):
                label = f"Total Number of people in the video {str(max(so_far))}"
            else:
                label = f"Total Number of people in the video {len(known_faces)}"
            cv2.putText(image, label, (17, 17), cv2.FONT_HERSHEY_SIMPLEX, np.round(w/2160,1)+0.1, (233, 130, 0), 1, cv2.LINE_AA)
            label = f"Number of people in the current frame {str(no_of_faces)}"
            cv2.putText(image, label, (w//2, 17), cv2.FONT_HERSHEY_SIMPLEX, np.round(w/2160,1)+0.1, (233, 130, 0), 1, cv2.LINE_AA)
            video.write(image)
            pre_frame = cur_frame
    cap.release()
    video.release()
    cv2.destroyAllWindows()


frame_queue = Queue()
darknet_image_queue = Queue(maxsize=1)
detections_queue = Queue(maxsize=1)
fps_queue = Queue(maxsize=1)
cords_que = Queue(maxsize=1)


network, class_names, class_colors = darknet.load_network("cfg/yolov4-obj.cfg", "data/obj.data", "yolov4-obj_best.weights",batch_size=1)


darknet_width = darknet.network_width(network)
darknet_height = darknet.network_height(network)


cap = cv2.VideoCapture('/content/test_7.mp4')
ret, img = cap.read()
[h,w] = img.shape[:2]
video = set_saved_video(cap, '/content/final.avi', (w, h))


t1 = Thread(target=video_capture, args=(frame_queue, darknet_image_queue))
t2 = Thread(target=inference, args=(darknet_image_queue, detections_queue, fps_queue))
t3 = Thread(target=drawing, args=(frame_queue, detections_queue, fps_queue))
# t4 = Thread(target=face_reg, args=(faces_que,img_queue))


t1.start()
t2.start()
t3.start()
# t4.start()


t1.join()
t2.join()
t3.join()
# t4.join()
```
